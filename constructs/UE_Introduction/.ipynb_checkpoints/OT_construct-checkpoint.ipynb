### 3- `Advanced I`: OT RNAP step size determination 

Read the following article:
[Direct observation of base-pair stepping by RNA polymerase](https://www.nature.com/articles/nature04268).  Here, we will be following their analysis protocol that led to the plots of Figure 2, detailed in the Methods section of the paper.

The position data was acquired at 2 kHz.  Load a file of experimental data (note: this data has been simulated, it is not from the paper).

````python
# load packages
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

#load data
data = np.loadtxt('OTsimdata.txt')
fs = 2000.
````

From the Methods section, we see that the authors first applied a 4th order, 1 kHz Bessel filter, then applied a median filter with a window length of 50 ms.

````python
# Bessel filter
order = 4
fc = 1000.
b, a = signal.bessel(order, fc/fs)
data_bessel = signal.lfilter(b, a, data)

# median filter
medfiltwin = 101  #(~= 50 ms, for a median filter, the window size must be odd)
data_bessel_median = signal.medfilt(data_bessel, medfiltwin)

# Plot the data
time = np.arange(len(data))/fs
plt.plot(time, data, label='raw data')
plt.plot(time, data_bessel, label='bessel filterd')
plt.plot(time, data_bessel_median, label='bessel and 50ms median filtered')
plt.ylabel('position ($\AA$)')
plt.xlabel('time (s)')
plt.legend()
````

The authors next made a histogram of the position data, using a histogram bin size of 0.1 angstrom. They then autocorrelated the histogram, calculated the power spectrum of the autocorrelation, then smoothed the power spectrum with a 5 point binomial filter.

````python
# calculate histogram
hist, bin_edges = np.histogram(data_bessel_median, bins = int(np.ptp(data_bessel_median)*10))

# autocorrelation of histogram
ac = np.correlate(hist, hist, mode='full')
ac = ac[ac.size//2:]
ac_dist = np.arange(len(ac))/ 10  # in angstroms -- remember, the bins of our hist were 0.1 angstrom

# power spectrum of autocorrelation
ps = np.abs(np.fft.fft(ac))**2
freqs = np.fft.fftfreq(ac.size, ac_dist[1])
idx = np.argsort(freqs)
freqs, ps = freqs[idx], ps[idx]

# smooth the power spectrum (don't worry too much about the details of the binomial filter)
def binomcoeffs(n): 
    return (np.poly1d([0.5, 0.5])**n).coeffs
ps_smoothed = np.convolve(ps, binomcoeffs(5),mode='same')

# some plots
fig, axes = plt.subplots(1, 3)
ax = axes.ravel()
fig.set_size_inches((20, 5))

ax[0].plot(bin_edges[0:-1]+bin_edges[1]/2,hist)
ax[0].set_ylabel('counts')
ax[0].set_xlabel('position ($\AA$)')
ax[0].set_title('histogram')

ax[1].plot(ac_dist[:200], ac[:200],'o-')
ax[1].set_xlabel('distance ($\AA$)')
ax[1].set_ylabel('autocorrelation')
ax[1].set_title('autocorrelation of hist')

pltidx = [int(len(freqs)/2), int(len(freqs)/2 + 80)]
ax[2].plot(freqs[pltidx[0]:pltidx[1]], ps[pltidx[0]:pltidx[1]], label='raw ps')
ax[2].plot(freqs[pltidx[0]:pltidx[1]], ps_smoothed[pltidx[0]:pltidx[1]], label='smoothed ps')
ax[2].set_xlabel('spatial frequency ($\AA^(-1)$)')
ax[2].set_ylabel('spectral density')
ax[2].set_ylim(0,np.max(ps_smoothed[pltidx[0]:pltidx[1]]))
plt.legend()
ax[2].set_title('power spectrum of autocorrelation')
````

### 4- `Advanced II`: OT simulation and stiffness callibration 

We've learned that a particle in an optical trap behaves as a particle in a harmonic potential.  Let's first simulate the Brownian motion of a particle in an optical trap.

````python
import numpy as np
import matplotlib.pyplot as plt

eta = 0.001   # [Pa*s] = [N*s/m^2]] water viscosity
dt = .05e-6   # time step of simulation, [s]
KT = 4.11e-21 # [Joule] = Boltzmann constant * temperature

def set_force(bead_xyz, stiff_ot):
    ''' calculate harmonic force from OT with a stiffness in (x,y,z) given in [N/m] '''
    force_ot = -stiff_ot * bead_xyz 
    return force_ot

def calc_gamma(bead_radius):
    '''calculates the friction coefficient [N s m^-1] of a spherical bead in water, given the radius [m]'''
    gamma0 = 6*np.pi*eta*bead_radius
    gamma = np.array([gamma0, gamma0, gamma0])
    return gamma

def make_trace(bead_radius, numb_points, stiff_ot):
    bead_xyz = np.zeros((numb_points, 3))
    noise = np.random.randn(numb_points, 3)
    gamma = calc_gamma(bead_radius)
    for i in range(1, numb_points):
        force = set_force(bead_xyz[i-1], stiff_ot)
        bead_xyz[i] = bead_xyz[i-1] + force * dt / gamma + np.sqrt(2*KT*dt/gamma)*noise[i-1]
    return bead_xyz

# Simulate a bead of 1 micron diameter. We'll simulate 1 million points, or 50ms.
bead_radius = 500e-9
num_pts = 1000000
stiff_ot = np.array([.1, .1,  5])*1e-3  #stiffness in (x, y, z) [N/m]
bead_xyz = make_trace(bead_radius, num_pts, stiff_ot)
time = np.arange(num_pts)*dt

# some plots
fig, axes = plt.subplots(1, 3)
ax = axes.ravel()
fig.set_size_inches((20, 5))

ax[0].plot(time, bead_xyz[:,0]*1e9,'-b',alpha=0.5, label='x')
ax[0].plot(time, bead_xyz[:,1]*1e9,'-r',alpha=0.5, label='y')
ax[0].set_ylabel('position (nm)')
ax[0].set_xlabel('time (s)')
ax[0].legend()

ax[1].plot(bead_xyz[:,0]*1e9, bead_xyz[:,1]*1e9, '.k', ms=.3, alpha=0.2)
ax[1].set_ylabel('y (nm)')
ax[1].set_xlabel('x (nm)')

ax[2].hist(bead_xyz[:,0]*1e9, 100, density=1, color='b', alpha=0.2, label='x')
ax[2].hist(bead_xyz[:,1]*1e9, 100, density=1, color='r', alpha=0.2, label='y')
ax[2].set_ylabel('Prob. density')
ax[2].set_xlabel('position (nm)')
ax[2].legend()
````

Try changing the trap stiffness in x and y using the stiff_ot variable. Then re-run the above block of code and see how the above plots change.

Now that we've simulated the 3D position of a particle in an optical trap, let's try two methods of OT calibration to make sure that we did our simulation correctly. Let's first use the Equipartition method.

````python
stiff_x = KT/np.var(bead_xyz[:,0])
stiff_y = KT/np.var(bead_xyz[:,1])
print('trap stiffness (x,y) by Equipartition method = (%.2e, %.2e) N/m'% (stiff_x, stiff_y))
````

You'll see that this is close to the trap stiffness given by the variable stiff_ot above.  You can change this value and re-run the code as a check.
Next, we'll calculate and plot the power spectral density of the x signal.  You could perform a fit of this signal to find the trap stiffness.  Here, we'll simply plot the theoretical Lorentzian curve.

````python
def make_spectrum(x, FPS):
    ''' Calculates the spectrum of x(t), and the log-avg spectrum. FPS = aquisition rate.'''
    spectrum = np.abs(np.fft.fft(x))**2
    freq = np.fft.fftfreq(len(x), d=1/FPS)
    freq = freq[:int(len(spectrum)/2.)]
    spectrum = spectrum[:int(len(spectrum)/2)]
    spectrum[1:] = 2*spectrum[1:]/(len(x)**2 * freq[1])
    return freq, spectrum

def Lorentzian(f, fc, drag):
    return KT/(np.pi**2*drag*(f**2+fc**2))

freqs, PSD_x = make_spectrum(bead_xyz[:,0], int(1/dt))
freqs, PSD_y = make_spectrum(bead_xyz[:,1], int(1/dt))
plt.loglog(freqs, PSD_x, '-b', alpha=0.3, label='x')
plt.loglog(freqs, PSD_y, '-r', alpha=0.3, label='y')

gamma = calc_gamma(bead_radius)
fc_x = stiff_ot[0]/(2*np.pi*gamma[0])
PSD_theory = Lorentzian(freqs, fc_x, gamma[0])
plt.loglog(freqs, PSD_theory, '-k',label='theory')
plt.ylabel('power spectrum')
plt.xlabel('freq (Hz)')
plt.legend()
````
